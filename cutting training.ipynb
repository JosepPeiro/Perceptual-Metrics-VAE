```python
import torchaudio
from preprocessing import Preprocessing
from torchaudio.transforms import Resample, MelSpectrogram, InverseMelScale, GriffinLim
import matplotlib.pyplot as plt
```

```python
audio_path = "./MusicCaps/_0-2meOf9qY.wav"

waveform, samp_rt = torchaudio.load(audio_path)

sample_rate = 16000
resampler = Resample(orig_freq=samp_rt , new_freq=sample_rate)
waveform = resampler(waveform)

print(f"Forma del waveform: {waveform.shape}")
print(f"Tasa de muestreo: {sample_rate} Hz")

primera_mitad = waveform[:,:(waveform.shape[1]//2)]
segunda_mitad = waveform[:,(waveform.shape[1]//2):]

crop_length = int(4.208 * sample_rate)
first_crop = primera_mitad[:,:crop_length]
second_crop = segunda_mitad[:,:crop_length]

second_crop.shape
```

```python
spectograma, maxi, mini = Preprocessing(first_crop, 16000, resampler_f = False)
print(spectograma.shape)
plt.imshow(spectograma,origin='lower')
```

```python
spectograma, maxi, mini = Preprocessing(second_crop, 16000, resampler_f = False)
print(spectograma.shape)
plt.imshow(spectograma,origin='lower')
```
