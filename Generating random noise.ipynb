```python
from preprocessing import Preprocessing, SplitAudio, Reconstruction
import torchaudio
import torch
```

```python
ruta_carpeta = "./MusicCaps/_0-2meOf9qY.wav"  # Reemplaza con la ruta real de tu carpeta

sample_rate_red = 16000
waveform, samp_rt = torchaudio.load(ruta_carpeta)
referencia, _ = SplitAudio(waveform, sample_rate = samp_rt, new_sample_rate = sample_rate_red)
ref, mx, mn = Preprocessing(referencia, 16000, resampler_f = False)
print(ref.max(), ref.min(), ref.shape)
print(type(ref))
```

```python
tensor_random = torch.rand(256, 256)
tensor_random
print(tensor_random.max(), tensor_random.min(), tensor_random.shape)
```

```python
from IPython.display import Audio
rec = Reconstruction(tensor_random, mx, mn)
Audio(rec.numpy(), rate=16000)
```
